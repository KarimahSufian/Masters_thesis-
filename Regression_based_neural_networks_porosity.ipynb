```
# Looking at the overall statistics of the data 
# calculating mean, standrard deviation etc. for each variable 
train_stats = train_dataset.describe()
train_stats.pop('PORO')
train_stats = train_stats.transpose()
train_stats
```

# Split features from labels

```
# splittig the features from labels
# seperate the target value/label from the features. 
# This label you specify here is the value that you want to train the model to predict 
train_labels = train_dataset.pop('PORO')
test_labels = test_dataset.pop('PORO')
```

# Normalize the data

We are normalizing features that use different scales and ranges. Without this step, it makes model training more difficult and makes the resulting model dependent on the choice of units used in the input. The ststistics (count, min, std etc.) genersted above are used to normalize the test dataset. This allows the test dataset to be projected into the same distribution that the model has been trained on

```
def norm(x):
  return (x - train_stats['mean']) / train_stats['std']
normed_train_data = norm(train_dataset)
normed_test_data = norm(test_dataset)
```

# Building the model

Let's build our model. Here, we'll use a Sequential model with two densely connected hidden layers, and an output layer that returns a single, continuous value. The model building steps are wrapped in a function, build_model, since we'll create a second model, later on.

```
from tensorflow.python import keras
from keras.layers import Dense, Activation
from keras.models import Sequential, Model
import numpy as np
```

```
def build_model():                                                                 # Defining a model
  model = keras.Sequential([                                                       # Using a sequential API model 
    layers.Dense(64, activation='elu', input_shape=[len(train_dataset.keys())]),  # first hidden layer contains 64 nodes, rectified linear unit activation function and a define input shape
    layers.Dense(64, activation='elu'),                                           # second hidden layer contains 64 nodes and rectified linear unit activation function
   layers.Dense(1)                                                                 # Output layer contains 1 node 
  ])

  optimizer = tf.keras.optimizers.RMSprop(0.001)                                  # Defining the optimizer: Using root mean squared [RMS] propogation optimizer, learning rate = 0.001
  #optimizer = tf.keras.optimizers.SGD(lr=0.01, momentum=0.9)
  model.compile(loss='mse',                                                       # compile model with loss = mean squared error
                optimizer=optimizer,                                              # optimizer as stated above 
                metrics=['mae', 'mse'])                                           # applying metrics 
  return model
```

```
model = build_model()
```

# Train the model

Train the model for 1000 epochs, and record the training and validation accuracy in the history object.

```
EPOCHS = 1000 # 1000 iterations set to predict and calculate for porosity 

history = model.fit(
  normed_train_data, train_labels,
  epochs=EPOCHS, validation_split = 0.2, verbose=0,     #validation_split = 0.2, 805 training data, 205 validation data 
  callbacks=[tfdocs.modeling.EpochDots()])

# mae: mean average estimate
# mse: mean squared estimate 
```

```
print (history.history.keys())
```

Visualize the model's training progress using the stats stored in the history object.

```
hist = pd.DataFrame(history.history)
hist['epoch'] = history.epoch
hist.head(999)
```

```
plotter = tfdocs.plots.HistoryPlotter(smoothing_std=2)
```

showing the change in predicted mean average estimate porosity (PORO) value against the number of iterations (EPOCHS). 
We can observe a convergence in the value for porosity

```
plt.figure(figsize=(15,5))
plotter.plot({'Basic': history}, metric = "loss")
plt.ylim([0, 12])
plt.ylabel('LOSS [PORO]')
```

```
plt.figure(figsize=(15,5))
plotter.plot({'Basic': history}, metric = "mae")
plt.ylim([0, 3])
plt.ylabel('MAE [PORO]')
```

```
plt.figure(figsize=(15,5))
plotter.plot({'Basic': history}, metric = "mse")
plt.ylim([0, 15])
plt.ylabel('MSE [PORO]')
```

This graph shows little improvement, or even degradation in the validation error after about 100 epochs. Let's update the model.fit call to automatically stop training when the validation score doesn't improve. We'll use an EarlyStopping callback that tests a training condition for every epoch. If a set amount of epochs elapses without showing improvement, then automatically stop the training.

```
model = build_model()

# The patience parameter is the amount of epochs to check for improvement
# after the porosity value stops varying for 20 iterations (patience = 20) then stop running the iterations
early_stop = keras.callbacks.EarlyStopping(monitor='val_loss', patience=20)

early_history = model.fit(normed_train_data, train_labels, 
                    epochs=EPOCHS, validation_split = 0.2, verbose=0, 
                    callbacks=[early_stop, tfdocs.modeling.EpochDots()])
```

```
plt.figure(figsize=(15,5))
plotter.plot({'Early Stopping': early_history}, metric = "loss")
plt.ylim([0, 15])
plt.ylabel('LOSS [PORO]')
plt.xlim([0, 64])
```

```
# plot porosity against number of iterations with the early cut off 
plt.figure(figsize=(15,5))
plotter.plot({'Early Stopping': early_history}, metric = "mae")
plt.ylim([0, 4])
plt.ylabel('MAE [PORO]')
plt.xlim([0, 64])
```

```
# plot porosity against number of iterations with the early cut off 
plt.figure(figsize=(15,5))
plotter.plot({'Early Stopping': early_history}, metric = "mse")
plt.ylim([0, 15])
plt.ylabel('MSE [PORO]')
plt.xlim([0, 64])
```

The graph above shows that on the validation set, the average error is usally around +/- 5 PORO. This is up to the user to decide if the model is good enough or not.

Let's see how well the model generalizes by using the test set, which we did not use when training the model. This tells us how well we can expect the model to predict when we use it in the real world.

```
loss, mae, mse = model.evaluate(normed_test_data, test_labels, verbose=2)

print("Testing set Mean Abs Error: {:5.5f} POR".format(mae))
```
